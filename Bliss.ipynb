# Модель Bliss для моделирования кривой процентных ставок

Модель Bliss (1997) - параметрическая модель для аппроксимации кривой доходности с использованием 5 параметров:
- β₀: уровень долгосрочной процентной ставки
- β₁: краткосрочный компонент
- β₂: среднесрочный компонент  
- τ₁: первый временной параметр
- τ₂: второй временной параметр

Формула: y(t) = β₀ + β₁ × [(1 - e^(-t/τ₁))/(t/τ₁)] + β₂ × [(1 - e^(-t/τ₂))/(t/τ₂) - e^(-t/τ₂)]

**Метод оптимизации:** `scipy.optimize.least_squares` с методом `'trf'` (Trust Region Reflective)


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import least_squares
import seaborn as sns

# Настройки для визуализации
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 6)
plt.rcParams['font.size'] = 10
```

```python
# Загрузка данных
df = pd.read_csv('data.csv')
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')
df.set_index('date', inplace=True)

print(f"Данные загружены: {df.shape[0]} наблюдений")
print(f"Сроки погашения (лет): {df.columns.tolist()}")
print(f"\nПериод данных: с {df.index[0].date()} по {df.index[-1].date()}")
df.head()
```

```python
class BlissModel:
    """
    Модель Bliss для моделирования кривой процентных ставок.
    
    Формула: y(t) = β₀ + β₁ × [(1 - e^(-t/τ₁))/(t/τ₁)] + β₂ × [(1 - e^(-t/τ₂))/(t/τ₂) - e^(-t/τ₂)]
    
    Параметры:
    - β₀ (beta0): уровень долгосрочной процентной ставки
    - β₁ (beta1): краткосрочный компонент
    - β₂ (beta2): среднесрочный компонент  
    - τ₁ (tau1): первый временной параметр
    - τ₂ (tau2): второй временной параметр
    """
    
    def __init__(self):
        self.params = None
        self.fitted = False
        self.optimization_result = None
        
    def yield_curve(self, t, beta0, beta1, beta2, tau1, tau2):
        """
        Вычисляет доходность для заданных сроков погашения.
        
        Parameters:
        -----------
        t : array-like
            Сроки погашения
        beta0, beta1, beta2, tau1, tau2 : float
            Параметры модели
            
        Returns:
        --------
        array : доходности для заданных сроков
        """
        t = np.array(t, dtype=float)
        
        # Избегаем деления на ноль для t ≈ 0
        eps = 1e-10
        t = np.where(t < eps, eps, t)
        
        # Первый компонент: [(1 - e^(-t/τ₁))/(t/τ₁)]
        ratio1 = t / tau1
        term1 = (1 - np.exp(-ratio1)) / ratio1
        
        # Второй компонент: [(1 - e^(-t/τ₂))/(t/τ₂) - e^(-t/τ₂)]
        ratio2 = t / tau2
        term2 = (1 - np.exp(-ratio2)) / ratio2 - np.exp(-ratio2)
        
        return beta0 + beta1 * term1 + beta2 * term2
    
    def residuals_function(self, params, t, observed_yields):
        """
        Функция остатков для least_squares.
        
        Parameters:
        -----------
        params : array-like
            [beta0, beta1, beta2, tau1, tau2]
        t : array-like
            Сроки погашения
        observed_yields : array-like
            Наблюденные доходности
            
        Returns:
        --------
        residuals : array
            Вектор остатков (observed - predicted)
        """
        beta0, beta1, beta2, tau1, tau2 = params
        predicted_yields = self.yield_curve(t, beta0, beta1, beta2, tau1, tau2)
        return observed_yields - predicted_yields
    
    def fit(self, t, yields, initial_params=None, method='trf'):
        """
        Обучает модель на данных.
        
        Parameters:
        -----------
        t : array-like
            Сроки погашения
        yields : array-like
            Наблюденные доходности
        initial_params : array-like, optional
            Начальные значения параметров [beta0, beta1, beta2, tau1, tau2]
        method : str, default='trf'
            Метод оптимизации для least_squares:
            - 'trf': Trust Region Reflective (рекомендуется, поддерживает bounds)
            - 'dogbox': Dogleg (также поддерживает bounds)
            - 'lm': Levenberg-Marquardt (не поддерживает bounds)
            
        Returns:
        --------
        self
        """
        if initial_params is None:
            # Начальные приближения
            beta0 = np.mean(yields[-3:])  # средняя долгосрочная ставка
            beta1 = yields[0] - beta0      # краткосрочная компонента
            beta2 = 0.0
            tau1 = 1.0  # начальное значение для первого временного параметра
            tau2 = 5.0  # начальное значение для второго временного параметра
            initial_params = [beta0, beta1, beta2, tau1, tau2]
        
        # Ограничения: τ₁ и τ₂ должны быть положительными
        # least_squares использует формат (lower, upper)
        lower_bounds = [-np.inf, -np.inf, -np.inf, 0.1, 0.1]
        upper_bounds = [np.inf, np.inf, np.inf, 10.0, 30.0]
        bounds = (lower_bounds, upper_bounds)
        
        # Оптимизация через least_squares (нелинейный МНК)
        result = least_squares(
            self.residuals_function,
            initial_params,
            args=(t, yields),
            method=method,
            bounds=bounds,
            ftol=1e-9,
            max_nfev=10000
        )
        
        self.params = result.x
        self.fitted = True
        self.optimization_result = result
        
        return self
    
    def predict(self, t):
        """
        Предсказывает доходности для заданных сроков погашения.
        
        Parameters:
        -----------
        t : array-like
            Сроки погашения
            
        Returns:
        --------
        array : предсказанные доходности
        """
        if not self.fitted:
            raise ValueError("Модель не обучена. Вызовите метод fit() сначала.")
        
        beta0, beta1, beta2, tau1, tau2 = self.params
        return self.yield_curve(t, beta0, beta1, beta2, tau1, tau2)
    
    def get_params_dict(self):
        """Возвращает параметры в виде словаря."""
        if not self.fitted:
            return None
        return {
            'β₀': self.params[0],
            'β₁': self.params[1],
            'β₂': self.params[2],
            'τ₁': self.params[3],
            'τ₂': self.params[4]
        }
    
    def score(self, t, yields):
        """
        Вычисляет R² и RMSE для оценки качества модели.
        
        Parameters:
        -----------
        t : array-like
            Сроки погашения
        yields : array-like
            Наблюденные доходности
            
        Returns:
        --------
        dict : метрики качества
        """
        if not self.fitted:
            raise ValueError("Модель не обучена.")
        
        predicted = self.predict(t)
        
        # R²
        ss_res = np.sum((yields - predicted) ** 2)
        ss_tot = np.sum((yields - np.mean(yields)) ** 2)
        r2 = 1 - (ss_res / ss_tot)
        
        # RMSE
        rmse = np.sqrt(np.mean((yields - predicted) ** 2))
        
        # MAE
        mae = np.mean(np.abs(yields - predicted))
        
        return {
            'R²': r2,
            'RMSE': rmse,
            'MAE': mae
        }

print("Класс BlissModel создан успешно!")
```

```python
# Обучение модели на одной дате для примера
# Возьмем первую дату из датасета

sample_date = df.index[0]
maturities = df.columns.astype(float).values
yields_sample = df.iloc[0].values

print(f"Обучение модели на дату: {sample_date.date()}")
print(f"Сроки погашения: {maturities}")
print(f"Процентные ставки: {yields_sample}")

# Создаем и обучаем модель
model = BlissModel()
model.fit(maturities, yields_sample, method='trf')

# Выводим результаты
print("\n" + "="*60)
print("РЕЗУЛЬТАТЫ ОПТИМИЗАЦИИ (least_squares)")
print("="*60)
print(f"Статус: {model.optimization_result.message}")
print(f"Количество вызовов функции: {model.optimization_result.nfev}")
print(f"Значение функции потерь (cost): {model.optimization_result.cost:.6f}")

print("\nПараметры модели:")
params_dict = model.get_params_dict()
for param, value in params_dict.items():
    print(f"  {param} = {value:.6f}")

# Оценка качества
metrics = model.score(maturities, yields_sample)
print("\nМетрики качества:")
for metric, value in metrics.items():
    print(f"  {metric} = {value:.6f}")
```

```python
# Визуализация результатов для одной даты

# Создаем более плотную сетку для сглаженной кривой
tau_fine = np.linspace(maturities.min(), maturities.max(), 300)
predicted_fine = model.predict(tau_fine)
predicted_original = model.predict(maturities)

plt.figure(figsize=(14, 7))

# График 1: Исходные и предсказанные значения
plt.subplot(1, 2, 1)
plt.plot(maturities, yields_sample, 'o', markersize=8, label='Наблюденные данные', color='darkblue')
plt.plot(tau_fine, predicted_fine, '-', linewidth=2, label='Модель Bliss', color='red')
plt.xlabel('Срок погашения (годы)', fontsize=12)
plt.ylabel('Процентная ставка (%)', fontsize=12)
plt.title(f'Кривая процентных ставок\n{sample_date.date()}', fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)

# График 2: Остатки (ошибки)
residuals = yields_sample - predicted_original
plt.subplot(1, 2, 2)
plt.bar(maturities, residuals, width=0.5, alpha=0.7, color='steelblue', edgecolor='black')
plt.axhline(y=0, color='red', linestyle='--', linewidth=2)
plt.xlabel('Срок погашения (годы)', fontsize=12)
plt.ylabel('Остатки (%)', fontsize=12)
plt.title('Остатки модели (наблюденное - предсказанное)', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"Максимальная ошибка: {np.max(np.abs(residuals)):.4f}%")
print(f"Средняя абсолютная ошибка: {np.mean(np.abs(residuals)):.4f}%")
```

## Обучение модели на всем датасете


```python
# Обучаем модель для каждой даты в датасете
from tqdm.auto import tqdm

n_dates = len(df)
results = {
    'date': [],
    'beta0': [],
    'beta1': [],
    'beta2': [],
    'tau1': [],
    'tau2': [],
    'rmse': [],
    'r2': [],
    'mae': [],
    'convergence': []
}

print(f"Обучение модели Bliss на {n_dates} датах...\n")

for date_idx in tqdm(range(n_dates), desc="Обучение"):
    date = df.index[date_idx]
    yields = df.iloc[date_idx].values
    
    # Создаем и обучаем модель
    model_temp = BlissModel()
    model_temp.fit(maturities, yields, method='trf')
    
    # Сохраняем результаты
    results['date'].append(date)
    results['beta0'].append(model_temp.params[0])
    results['beta1'].append(model_temp.params[1])
    results['beta2'].append(model_temp.params[2])
    results['tau1'].append(model_temp.params[3])
    results['tau2'].append(model_temp.params[4])
    
    # Метрики
    metrics = model_temp.score(maturities, yields)
    results['rmse'].append(metrics['RMSE'])
    results['r2'].append(metrics['R²'])
    results['mae'].append(metrics['MAE'])
    results['convergence'].append(model_temp.optimization_result.success)

# Создаем DataFrame с результатами
results_df = pd.DataFrame(results)
results_df.set_index('date', inplace=True)

print(f"\n✓ Модель успешно обучена на всех датах!")
print(f"✓ Успешная сходимость: {results_df['convergence'].sum()}/{n_dates} ({100*results_df['convergence'].mean():.1f}%)")
print(f"\nСредние метрики качества:")
print(f"  RMSE: {results_df['rmse'].mean():.4f}")
print(f"  R²: {results_df['r2'].mean():.4f}")
print(f"  MAE: {results_df['mae'].mean():.4f}")
```

## Заключение

**Реализована модель Bliss для моделирования кривой процентных ставок:**

**Модель:**
- Формула: y(t) = β₀ + β₁ × [(1 - e^(-t/τ₁))/(t/τ₁)] + β₂ × [(1 - e^(-t/τ₂))/(t/τ₂) - e^(-t/τ₂)]
- 5 параметров: β₀ (уровень), β₁ (краткосрочный), β₂ (среднесрочный), τ₁ и τ₂ (временные параметры)

**Метод обучения:**
- **least_squares** с методом **'trf'** (Trust Region Reflective)
- Нелинейный метод наименьших квадратов (Levenberg-Marquardt family)
- Целевая функция: минимизация суммы квадратов остатков
- Ограничения: 0.1 < τ₁ < 10.0 и 0.1 < τ₂ < 30.0

**Результаты:**
- Модель успешно обучена на всех датах датасета
- Высокое качество аппроксимации кривой доходности (высокий R², низкие RMSE и MAE)
- Параметры модели изменяются во времени, отражая динамику процентных ставок
- Два временных параметра позволяют лучше моделировать различные формы кривой

**Применение:**
- Прогнозирование процентных ставок для любых сроков погашения
- Анализ формы кривой доходности (нормальная, инвертированная, плоская)
- Оценка премий за срочность
- Управление процентным риском
- Ценообразование облигаций и процентных производных инструментов
